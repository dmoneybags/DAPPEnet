```python
import json
import pandas as pd
import os
import pprint
```

```python
class Databases:
    namesWithDash = ["Kommo-o", "Ho-oh", "Porygon-Z", "Tapu Koko", "Tapu Lele", 
                     "Tapu Bulu", "Tapu Fini"]
    pokemonTypes = {}
    moves = {}
    items = {}
    statuses = {}
    conditions = {}
    sideEffects = {}
    weather = {}
    terrain = {}
    def getCanonicalName(name):
        #if '-' in name and name != Databases.namesWithDash:
            #return name.split('-')[0]
        return name
    def cleanVal(val):
        val = val.strip()
        val = val.lower()
        val = val.replace(" ", "")
        return val
    def load():
        Databases.loadPkmn()
        Databases.loadItems()
        Databases.loadMoves()
        Databases.loadStatuses()
        Databases.conditions = pd.read_excel(os.getcwd() + '/Databases/conditions.xlsx')
        Databases.sideEffects = pd.read_excel(os.getcwd() + '/Databases/sideeffects.xlsx')
        Databases.weather = pd.read_excel(os.getcwd() + '/Databases/weather.xlsx')
        Databases.terrain = pd.read_excel(os.getcwd() + '/Databases/terrain.xlsx')
    def loadPkmn():
        df = pd.read_excel(os.getcwd() + '/Databases/pkmntypes.xlsx')
        df['Name'] = df['Name'].apply(str.lower)
        df = df.fillna(0)
        Databases.pokemonTypes = df
    def loadItems():
        df = pd.read_csv(os.getcwd() + '/Databases/items.csv')
        df['Name'] = df['Name'].apply(str.lower)
        df['Name'] = df['Name'].str.strip()
        df['Name'] = df['Name'].str.replace(" ", "")
        df = df.fillna(0)
        Databases.items = df
    def loadMoves():
        df = pd.read_csv(os.getcwd() + '/Databases/movesdata.csv')
        df['Name'] = df['Name'].apply(str.lower)
        df['Name'] = df['Name'].str.strip()
        df['Name'] = df['Name'].str.replace(" ", "")
        df['Name'] = df['Name'].str.replace("-", "")
        df['Name'] = df['Name'].str.replace("*", "", regex=False)
        df = df.fillna(0)
        Databases.moves = df
    def loadStatuses():
        df = pd.read_csv(os.getcwd() + '/Databases/status.csv')
        df['Name'] = df['Name'].apply(str.lower)
        Databases.statuses = df
    def processInputPkmn(pkmn):
        #if '-' in pkmn and pkmn not in Databases.namesWithDash:
            #return Databases.getCanonicalName(pkmn) + " (" + pkmn + ")"
        return pkmn
    def getPkmnTypes(pkmn):
        print(pkmn)
        pkmn = Databases.processInputPkmn(pkmn)
        pkmn = pkmn.lower()
        print(pkmn)
        pkmnDataRow = Databases.pokemonTypes.loc[Databases.pokemonTypes['Name'] == pkmn]
        type1 = pkmnDataRow['Type 1']
        type2 = pkmnDataRow['Type 2']
        print(pkmnDataRow)
        return int(type1), int(type2)
    def getItemNum(item):
        item = item.lower()
        item = item.replace(" ", "")
        itemRow = Databases.items.loc[Databases.items['Name'] == item]
        return itemRow['Num']
    def getMoveList(move):
        move = Databases.cleanVal(move)
        print(move)
        moveRow = Databases.moves.loc[Databases.moves['Name'] == move]
        print(moveRow)
        return [float(moveRow['Power']), float(moveRow['Type']), float(moveRow['Accuracy'])]
    def getVal(val, df):
        if val == '':
            return 0
        val = Databases.cleanVal(val)
        names = df['Name'].tolist()
        return names.index(val) + 1
```
